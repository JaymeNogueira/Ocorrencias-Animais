# Ocorrencia com Todos os Anfibios

Nesse Notebook trabalharemos somente com as ocorrencias que estiverem associadas aos atropelamento de Anfibios e Faixas de Kilometros.
Iniciaremos importando os dados ja corrigidos previamente e as bibliotecas.

```python
# data analysis and wrangling
import pandas as pd
import numpy as np
import random as rnd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

dc = pd.read_csv('dadoscorrigidos.csv')
tbanimal = pd.read_csv('animal.csv')
```

```python
dc.head(10)
```

Agora removeremos todos os atributos que não forem relevantes para as ocorrencias com os Anfibios.

```python
anfdc = dc.drop(['Taxonomia','codAnimal', 'Mamifero', 'Reptil', 'Ave'], axis=1)
anfdc.head()
```

Em seguida é demonstrado o grafico de correlação de atributos de pearson

```python
colormap = plt.cm.viridis
plt.figure(figsize=(12,12))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(anfdc.astype(float).corr(),linewidths=0.1,vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True)
```

Em seguida alteramos os dados para que cada dado de uma ocorrencia esteja associada a seu respectivo atributo.

```python
anfasc = anfdc
headers = ['FaixaKm','TipoPavimento','TipoChuva','Intervencao',
           'Vazamento','agua','vegetacaoBaixa','capimAlto','arbustos','floresta','TipoLocal','Anfibio']
for i in range(len(headers)):
    anfasc[headers[i]] = headers[i]+'='+anfasc[headers[i]].apply(str)

anfasc.head()
```

Enfim aplicaremos o Apriori usando suporte minimo de 0.03 e confianca de 0.8

```python
listanfasc = []
for i in range(len(anfasc)):
    listanfasc.append([str(anfasc.values[i, j]) for j in range(0, 12)])

from apyori import apriori
regras = apriori(listanfasc, min_support=0.03, min_confidence=0.8, min_lenght=2)

resultados = list(regras)
resultados2 = [list(x) for x in resultados]
resultadosFormatado = []
for j in range(0,3000):#numero de regras que deseja mostrar
       resultadosFormatado.append([list(x) for x in resultados2[j][2]])
resultdf = pd.DataFrame(resultadosFormatado)
resultdf.to_csv('anfregras.csv', index=False, header=False)
```

Porem por mais que aumentemos a quantidade de regras não foi possivel achar nenhuma associação entre as ocorrencias com Anfibios e as Faixas de Km

# Ocorrencias com o Anfibio mais Atropelado

Para trabalharmos somente com os Anfibios mais atropelados precisamos primeiro remover as ocorrencias com as outras taxonomias.

```python
anfdc = dc.drop(['Taxonomia', 'Mamifero', 'Reptil', 'Ave'], axis=1)
anfdc = anfdc[anfdc.Anfibio == 1]
anfdc.head()
```

Em seguida selecionaremos os animais com maiores numeros de ocorrencia

```python
anfdc.groupby('codAnimal').size().sort_values(ascending=False).head(3)
```

Agora iremos na tabela dos animais identificar qual é o animal mais atropelado

```python
tbanimal[tbanimal.codAnimal == 307]
```

Agora que ja identificamos o nome do animal mais atropelado, criaremos uma tributo novo separando qual é a ave mais atropelada e qual não é

```python
anfdc['Sapo'] = (anfdc['codAnimal'] == 307)*1
anfdc.head()
```

Correlação de Pearson para os atributos e o Anú-Preto

```python
sapodc = anfdc.drop(['codAnimal', 'Anfibio'], axis=1)
colormap = plt.cm.viridis
plt.figure(figsize=(12,12))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(sapodc.astype(float).corr(),linewidths=0.1,vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True)
```

Fazendo a mesma associação entre os dados e os atributos

```python
headers = ['FaixaKm','TipoPavimento','TipoChuva','Intervencao',
           'Vazamento','agua','vegetacaoBaixa','capimAlto','arbustos','floresta','TipoLocal','Sapo']
for i in range(len(headers)):
    sapodc[headers[i]] = headers[i]+'='+sapodc[headers[i]].apply(str)

sapodc.head()
```

```python
listsapo = []
for i in range(len(sapodc)):
    listsapo.append([str(sapodc.values[i, j]) for j in range(0, 12)])

from apyori import apriori
regras = apriori(listsapo, min_support=0.03, min_confidence=0.8, min_lenght=2)

resultados = list(regras)
resultados2 = [list(x) for x in resultados]
resultadosFormatado = []
for j in range(0,3000):#numero de regras que deseja mostrar
       resultadosFormatado.append([list(x) for x in resultados2[j][2]])
resultdf = pd.DataFrame(resultadosFormatado)
resultdf.to_csv('saporegras.csv', index=False, header=False)
```
